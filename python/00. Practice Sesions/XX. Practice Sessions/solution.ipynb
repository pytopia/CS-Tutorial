```python
from collections import defaultdict, Counter
```

```python
mylist = list(range(1000))
```

```python
%%timeit
# solution 1: O(n) o(n)
counter = {}
for item in mylist:
    if item in counter:
        counter[item] += 1
    else:
        counter[item] = 1
```

```python
%%timeit
# solution 2: O(n) o(2n)
counter = {}
for item in mylist:
    counter[item] = counter.get(item, 0) + 1
```

```python
%%timeit
# solution 3: O(n) o(3n)
counter = defaultdict(int)
for item in mylist:
    counter[item] += 1
```

```python
%%timeit
# solution 4: O(n) o(0.5n)
Counter(mylist)
```

```python
%%timeit
# solution 5: O(n^2)
counter = {}
for item in set(mylist):
    counter[item] = mylist.count(item)
```
