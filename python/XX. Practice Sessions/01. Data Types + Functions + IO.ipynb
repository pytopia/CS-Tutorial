### 9. Decode Nazis Messages

German Nazis are using a system (Enigma) to encode their messages through transmission. We know that the system shifts every character with a constant number and all chracters are ASCII. For example with a `shift = 2`: "abcde" is encrypted to "cdefg"

We also know that Germans put the phrase "Heil Hitler" in every encrypted message.

1. Write a function that can decode a German message provided a constant shift. Write the decoded message in a `decoded.txt` file.
2. What is the time and space complexity of decode function? what about your final code?

````python
with open("files/encoded.txt") as f:
    content = f.read()
````

````python
def decode_by_shift(message, shift):
    """
    Decode a message by a given shift.
    """
    decoded_msg = ""
    for char in message:
        decoded_char = chr((ord(char) - shift) % 128)
        decoded_msg += decoded_char
    return decoded_msg
````

````python
def decode(message):
    """
    Decode a message by trying different shifts.
    """
    correct_shift = None
    for shift in range(128):
        decoded_msg = decode_by_shift(content, shift)

        # check if flag exists in the code
        if "Heil Hitler" in decoded_msg:
            correct_shift = shift
            
    return decode_by_shift(content, correct_shift)
````

> **Solution**

```python
with open("files/encoded.txt") as f:
    content = f.read()
    
def decode(message, shift):
    decoded_msg = ""
    for char in message:
        decoded_msg += chr((ord(char) - shift) % 128)
        
    return decoded_msg

correct_shift = None
for shift in range(128):
    decoded_msg = decode(content, shift)
    if "Heil Hitler" in decoded_msg:
        correct_shift = shift
        
if correct_shift:
    decoded_msg = decode(content, correct_shift)
    print(f"Decoded the message successfuly. (shift = {correct_shift})")
    print("-"*50)
    print("Decoded message:")
    print(decoded_msg)
else:
    print("Could not decode the message.")
```
